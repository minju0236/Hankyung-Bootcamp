```python
%%writefile test.c
#include <stdio.h>
#include <string.h>
#include <stdlib.h>

void print_help() {
    printf("사용법: program [옵션] [인자]\n\n");
    printf("옵션:\n");
    printf("  -h, --help          도움말 출력\n");
    printf("  -v, --version       버전 정보 출력\n");
    printf("  -f, --file <파일>   파일 이름 지정\n");
    printf("  -n, --number <숫자> 숫자 값 지정\n");
    printf("  -l, --list          목록 모드\n");
    printf("  -d, --debug         디버그 모드\n");
    printf("\n예제:\n");
    printf("  ./program -f test.txt -n 10\n");
    printf("  ./program --debug --list\n");
}

void print_version() {
    printf("프로그램 버전 1.0.0\n");
    printf("Copyright (c) 2024\n");
}

int main(int argc, char *argv[]) {
    // 옵션 플래그
    int debug_mode = 0;
    int list_mode = 0;
    char *filename = NULL;
    int number = 0;

    printf("=== CLI 명령어 파서 예제 ===\n\n");

    // argc와 argv 정보 출력
    printf("[프로그램 정보]\n");
    printf("프로그램 이름: %s\n", argv[0]);
    printf("전체 인자 개수: %d\n\n", argc);

    // 모든 인자 출력
    printf("[받은 인자 목록]\n");
    for (int i = 0; i < argc; i++) {
        printf("argv[%d] = \"%s\"\n", i, argv[i]);
    }
    printf("\n");

    // 인자가 없으면 도움말 출력
    if (argc == 1) {
        print_help();
        return 0;
    }

    // 옵션 파싱
    printf("[옵션 파싱 결과]\n");
    for (int i = 1; i < argc; i++) {
        // 도움말
        if (strcmp(argv[i], "-h") == 0 || strcmp(argv[i], "--help") == 0) {
            print_help();
            return 0;
        }
        // 버전
        else if (strcmp(argv[i], "-v") == 0 || strcmp(argv[i], "--version") == 0) {
            print_version();
            return 0;
        }
        // 디버그 모드
        else if (strcmp(argv[i], "-d") == 0 || strcmp(argv[i], "--debug") == 0) {
            debug_mode = 1;
            printf("✓ 디버그 모드 활성화\n");
        }
        // 목록 모드
        else if (strcmp(argv[i], "-l") == 0 || strcmp(argv[i], "--list") == 0) {
            list_mode = 1;
            printf("✓ 목록 모드 활성화\n");
        }
        // 파일 옵션
        else if (strcmp(argv[i], "-f") == 0 || strcmp(argv[i], "--file") == 0) {
            if (i + 1 < argc) {
                filename = argv[i + 1];
                printf("✓ 파일 지정: %s\n", filename);
                i++; // 다음 인자는 이미 처리했으므로 건너뛰기
            } else {
                printf("✗ 오류: -f 옵션에 파일명이 필요합니다.\n");
                return 1;
            }
        }
        // 숫자 옵션
        else if (strcmp(argv[i], "-n") == 0 || strcmp(argv[i], "--number") == 0) {
            if (i + 1 < argc) {
                number = atoi(argv[i + 1]);
                printf("✓ 숫자 지정: %d\n", number);
                i++;
            } else {
                printf("✗ 오류: -n 옵션에 숫자가 필요합니다.\n");
                return 1;
            }
        }
        // 알 수 없는 옵션
        else {
            printf("✗ 알 수 없는 옵션: %s\n", argv[i]);
            printf("   -h 또는 --help로 도움말을 확인하세요.\n");
            return 1;
        }
    }

    // 실행 결과
    printf("\n[실행 결과]\n");

    if (debug_mode) {
        printf("[DEBUG] 디버그 정보를 출력합니다.\n");
    }

    if (list_mode) {
        printf("목록 항목:\n");
        printf("  1. 첫 번째 항목\n");
        printf("  2. 두 번째 항목\n");
        printf("  3. 세 번째 항목\n");
    }

    if (filename != NULL) {
        printf("파일 '%s'를 처리합니다...\n", filename);
        if (debug_mode) {
            printf("[DEBUG] 파일 경로: %s\n", filename);
        }
    }

    if (number != 0) {
        printf("숫자 연산 결과:\n");
        printf("  입력값: %d\n", number);
        printf("  2배: %d\n", number * 2);
        printf("  제곱: %d\n", number * number);
    }

    printf("\n프로그램 실행 완료!\n");

    return 0;
}
```

# Cloudflare Tunnel을 사용하여 node코드 공인 ip없이 url 생성해서 접속하기

```python
!npm init -y
!npm install express
```

```python
%%writefile server.js
const express = require("express");
const app = express();

app.get("/", (req, res) => {
  res.send("HTTP Server Running");
});

app.listen(3000, () => {
  console.log("Server running on port 3000");
});
```

node server.js를 백그라운드에서 계속 실행하고, 실행 로그(표준출력 + 에러출력)를 server.log 파일에 저장하는 명령어

2>&1의미: 에러 출력(2)을 표준 출력(1)과 같은 곳으로 보내라 (>은 보낸다, &은 통로)

```python
!nohup node server.js > server.log 2>&1 &
```

3000포트 실행

```python
!tail -n 10 server.log
```

cloudflare 설치


```python
!npm install -g cloudflared
```

Cloudflare Tunnel을 백그라운드에서 실행해서 localhost:3000을 외부에서 접속 가능하게 열고, 실행 로그를 tunnel.log 파일에 저장하는 명령어


```python
!nohup cloudflared tunnel --url http://localhost:3000 > tunnel.log 2>&1 &
```

tunnel.log 파일에서 생성된 trycloudflare.com 공개 URL만 추출해서 가장 마지막(최신) 주소 하나만 출력하는 명령어

```python
!grep -o "https://[a-zA-Z0-9.-]*\.trycloudflare.com" tunnel.log | tail -n 1
```
