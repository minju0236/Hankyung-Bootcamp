```python
%%writefile ext.js

// 1. const: 변하지 않는 값
const PI = 3.14159;
const myName = "Gemini";
// myName = "AI"; // 에러 발생: Assignment to constant variable.

// 2. let: 변하는 값
let score = 0;
score = 10; // 재할당 가능
console.log("현재 점수:", score);

// 3. const와 객체 (참조 타입의 특이점)
const user = { id: 1, name: "철수" };
user.name = "영희"; // 객체 내부의 속성은 변경 가능 (메모리 주소는 그대로이기 때문)
console.log("사용자 정보:", user);
```

```python
%%writefile conditional.js

// 1. if...else 문
const age = 21;
if (age >= 20) {
    console.log("성인입니다.");
} else {
    console.log("미성년자입니다.");
}

// 2. switch 문 (요일 확인)
const day = "월";
switch (day) {
    case "월":
        console.log("월요병 극복 중...");
        break;
    case "금":
        console.log("불타는 금요일!");
        break;
    default:
        console.log("평범한 하루입니다.");
}

// 3. 삼항 연산자와 엄격한 일치(===)
const input = "10";
const message = (input === 10) ? "숫자 10입니다." : "숫자 10이 아니거나 타입이 다릅니다.";
console.log(message); // "10"은 문자열이므로 거짓 출력
```

```python
!node conditional.js
```

```python
%%writefile for.js

// 1. 기본 for문: 1부터 5까지 출력
console.log("--- for문 시작 ---");
for (let i = 1; i <= 5; i++) {
    console.log(`현재 번호는 ${i}입니다.`);
}

// 2. while문: 조건이 충족될 때까지 반복
let energy = 3;
console.log("--- while문 시작 ---");
while (energy > 0) {
    console.log(`에너지 소모 중... 남은 에너지: ${energy}`);
    energy--;
}
console.log("에너지가 방전되었습니다.");

// 3. for...of문: 배열 데이터 순회
const fruits = ["사과", "바나나", "포도"];
console.log("--- for...of문 시작 ---");
for (const fruit of fruits) {
    console.log(`내가 좋아하는 과일: ${fruit}`);
}
```

```python
!node for.js
```

```python
%%writefile function.js

// 1. 함수 표현식: 두 수의 합을 구함
const addNumbers = function(a, b) {
    return a + b;
};
const sum = addNumbers(10, 20);
console.log("합계:", sum);

// 2. 화살표 함수: 인사 메시지 생성
// 로직이 한 줄일 때는 { }와 return을 생략할 수 있습니다.
const getGreeting = (userName) => `안녕하세요, ${userName}님! 오늘 하루도 화이팅!`;
console.log(getGreeting("철수"));

// 3. 함수를 인자로 넘기는 콜백 체험 (일급 객체)
const runner = (action) => {
    console.log("달리기를 시작하기 전에...");
    action(); // 인자로 받은 함수를 실행
};

runner(() => console.log("열심히 달립니다! 🏃‍♂️"));
```

```python
!node function.js
```

```python
%%writefile ex3.js

// 1. let 재선언 시도 (에러 발생 확인)
let studentId = 20240001;
// let studentId = 20249999; // 주석을 해제하고 실행해 보세요. 에러가 발생합니다.

// 2. TDZ(일시적 사각지대) 체험
try {
    console.log(age); // ReferenceError 발생
    let age = 20;
} catch (e) {
    console.log("에러 발생:", e.message);
}

// 3. const 재할당 시도
const university = "한국대학교";
try {
    university = "서울대학교"; // 재할당 시도
} catch (e) {
    console.log("상수 변경 에러:", e.message);
}
```

```python
%%writefile ex4.js

{
    var globalScope = "저는 var라서 밖에서도 보여요";
    let localScope = "저는 let이라서 이 안에서만 살아요";
    console.log("블록 내부:", localScope);
}

console.log("블록 외부(var):", globalScope);

try {
    console.log("블록 외부(let):", localScope);
} catch (e) {
    console.log("블록 외부(let) 접근 결과: 에러 발생!");
}
```

```python
!node ex4.js
```
