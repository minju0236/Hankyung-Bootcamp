```python
%%writefile server.js

const express = require('express');
const path = require('path');
const mysql = require('mysql2');
const app = express();
const PORT = 3000;

// [1] form 데이터 처리 설정
app.use(express.urlencoded({ extended: true }));

// [2] DB 연결 설정
const conn = mysql.createConnection({
    host: 'localhost',
    user: 'newuser',
    password: '5678',
    database: 'testdb'
});

conn.connect((err) => {
    if (err) {
        console.error('DB 연결 실패:', err);
        return;
    }
    console.log('DB 연결 성공');
});

// [3] 회원가입 페이지 출력
app.get('/', (req, res) => {
    res.sendFile(path.join(__dirname, 'templates', 'signup.html'));
});

// [4] 회원가입 처리
app.post('/signup', (req, res) => {
    const { email, name, auth, dept, tel, rank, phone, memo } = req.body;

    const sql = `
        INSERT INTO users (email, name, auth, dept, tel, rank, phone, memo)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    `;

    const values = [ email, name, auth, dept, tel, rank, phone, memo];

    conn.query(sql, values, (err, result) => {
        if (err) {
            console.error('회원가입 INSERT 실패:', err);
            return res.send(`
                <h1>회원가입 실패</h1>
                <p>데이터 저장 중 오류가 발생했습니다.</p>
                <a href="/">회원가입 페이지로 돌아가기</a>
            `);
        }

        console.log('회원가입 성공:', result);
        res.sendFile(path.join(__dirname, 'templates', 'signup-success.html'));
    });
});

app.listen(PORT, () => {
    console.log('=========================================');
    console.log(` 회원가입 서버가 포트 ${PORT}에서 작동 중입니다.`);
    console.log('=========================================');
});
```
