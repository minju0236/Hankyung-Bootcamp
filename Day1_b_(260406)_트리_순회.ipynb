트리 - 전위 탐색

*   시간복잡도: O(logn)
*   탐색순서: 부모 -> 왼쪽 -> 오른쪽



```python
%%writefile tree.js

function createNode(value) { // 노드 생성 함수
  return {
    value: value,
    left: null,
    right: null
  };
}

// 노드 생성 코드
const root = createNode('A');

root.left = createNode('B');
root.right = createNode('C');

root.left.left = createNode('D');
root.left.right = createNode('E');

function printHorizontal(node) { // 노드 형태 출력 함수
  console.log("--- Tree Structure ---");
  console.log("        " + node.value);
  console.log("      /   \\");
  console.log("     " + node.left.value + "     " + node.right.value);
  console.log("    / \\");
  console.log("   " + node.left.left.value + "   " + node.left.right.value);
}

//위에서 사용한 preorder 함수는 결과를 저장하는 내용이 없기 때문에 결과 저장을 추가
function getPreorder(node, result = []) { // // 전위 탐색 (루트-왼-오)
  if (!node) return result;
  result.push(node.value); // // 탐색한 루트 저장
  getPreorder(node.left, result); // // 그 다음 왼쪽 노드 탐색
  getPreorder(node.right, result); // // 마지막으로 오른쪽 노드 탐색
  return result;
}

printHorizontal(root); // 노드 형태 출력
console.log("\n--- Preorder Traversal ---");
const traversalResult = getPreorder(root); // 탐색한 노드 값
console.log(traversalResult.join(" -> ")); // 탐색 순서 출력
```

트리 - 중위 탐색
*   시간복잡도: O(logn)
*   탐색순서: 왼쪽 -> 부모 -> 오른쪽


```python
%%writefile tree.js

function createNode(value) { // 노드 생성 함수
  return {
    value: value,
    left: null,
    right: null
  };
}

// 노드 생성 코드
const root = createNode('A');

root.left = createNode('B');
root.right = createNode('C');

root.left.left = createNode('D');
root.left.right = createNode('E');

function printHorizontal(node) { // 노드 형태 출력 함수
  console.log("--- Tree Structure ---");
  console.log("        " + node.value);
  console.log("      /   \\");
  console.log("     " + node.left.value + "     " + node.right.value);
  console.log("    / \\");
  console.log("   " + node.left.left.value + "   " + node.left.right.value);
}

function getInorder(node, result = []) { // 중위 탐색 (왼-루트-오)
  if (!node) return result;
  getInorder(node.left, result); // 왼쪽부터 탐색
  result.push(node.value); // 탐색한 노드 기록
  getInorder(node.right, result); // 마지막으로 오른쪽 탐색
  return result;
}

printHorizontal(root);
console.log("\n--- Inorder Traversal ---");
const traversalResult = getInorder(root); // 탐색한 노드 값
console.log(traversalResult.join(" -> ")); // 탐색 순서 출력
```

트리 - 후위 탐색
*   시간복잡도: O(logn)
*   탐색순서: 왼쪽 -> 오른쪽 -> 부모

```python
%%writefile tree.js

function createNode(value) { // 노드 생성 함수
  return {
    value: value,
    left: null,
    right: null
  };
}

// 노드 생성 코드
const root = createNode('A');

root.left = createNode('B');
root.right = createNode('C');

root.left.left = createNode('D');
root.left.right = createNode('E');

function printHorizontal(node) { // 노드 형태 출력 함수
  console.log("--- Tree Structure ---");
  console.log("        " + node.value);
  console.log("      /   \\");
  console.log("     " + node.left.value + "     " + node.right.value);
  console.log("    / \\");
  console.log("   " + node.left.left.value + "   " + node.left.right.value);
}

function getPostorder(node, result = []) { // 후위 탐색 (왼-오-루트)
  if (!node) return result;
  // Left -> Right -> Root
  getPostorder(node.left, result); // 왼쪽부터 탐색
  getPostorder(node.right, result); // 그 다음 오른쪽 탐색
  result.push(node.value); // 마지막으로 탐색한 노드 기록
  return result;
}

printHorizontal(root);
console.log("\n--- Postorder Traversal ---");
const traversalResult = getPostorder(root); // 탐색한 노드 값
console.log(traversalResult.join(" -> ")); // 탐색 순서 출력
```

```python
%%writefile insert.js

// 새로운 노드 객체를 생성하는 함수
function createNode(value) {
  return {
    value: value,
    left: null,
    right: null
  };
}

// 레벨 순서 삽입 (BFS) 방식: 빈 자리를 왼쪽부터 차례대로 채움
function insert(root, value) {
  const newNode = createNode(value);
  // 루트가 없으면 새 노드가 루트가 됨
  if (!root) return newNode;

  // 큐를 이용하여 트리 노드를 순회
  let queue = [root];
  while (queue.length > 0) {
    let node = queue.shift();

    // 왼쪽 자식이 없으면 삽입 후 종료
    if (!node.left) {
      node.left = newNode;
      return root;
    } else {
      // 있으면 다음에 탐색하기 위해 큐에 추가
      queue.push(node.left);
    }

    // 오른쪽 자식이 없으면 삽입 후 종료
    if (!node.right) {
      node.right = newNode;
      return root;
    } else {
      // 있으면 다음에 탐색하기 위해 큐에 추가
      queue.push(node.right);
    }
  }
  return root;
}

// 초기 트리 구축 (A를 루트로 설정)
let root = createNode('A');
insert(root, 'B');
insert(root, 'C');
insert(root, 'D');
insert(root, 'E');
insert(root, 'F');

// 트리의 구조를 콘솔에 아스키 아트로 출력하는 함수
function printHorizontal(node) {
  console.log("--- Tree Structure ---");
  console.log("        " + node.value);
  console.log("      /   \\");
  console.log("     " + node.left.value + "     " + node.right.value);
  console.log("    / \\   / ");
  // 하위 노드 존재 여부를 체크하며 출력
  console.log("   " + node.left.left.value + "   " + node.left.right.value + " " + (node.right.left ? node.right.left.value : " "));
}

// 최종 트리 상태 출력
printHorizontal(root);
```

```python
%%writefile delete.js

// 새로운 노드 객체를 생성하는 함수
function createNode(value) {
  return {
    value: value,
    left: null,
    right: null
  };
}

// 리프 노드를 삭제하는 함수
function deleteLeaf(node, target) {
  if (!node) return null;

  if (
    node.left &&
    node.left.value === target &&
    !node.left.left &&
    !node.left.right
  ) {
    node.left = null;   // 연결 제거
    return node;
  }

  if (
    node.right &&
    node.right.value === target &&
    !node.right.left &&
    !node.right.right
  ) {
    node.right = null;
    return node;
  }

  deleteLeaf(node.left, target);
  deleteLeaf(node.right, target);

  return node;
}

// 트리의 구조를 콘솔에 아스키 아트로 출력하는 함수
function printHorizontal(node) {
  if (!node) return;
  console.log("--- Tree Structure ---");
  console.log("        " + node.value);
  console.log("      /   \\");

  let leftVal = node.left ? node.left.value : " ";
  let rightVal = node.right ? node.right.value : " ";
  console.log("     " + leftVal + "     " + rightVal);
  console.log("    / \\   / ");

  let ll = (node.left && node.left.left) ? node.left.left.value : " ";
  let lr = (node.left && node.left.right) ? node.left.right.value : " ";
  let rl = (node.right && node.right.left) ? node.right.left.value : " ";

  console.log("   " + ll + "   " + lr + " " + rl);
}

let root = createNode('A');
root.left = createNode('B');
root.right = createNode('C');
root.left.left = createNode('D');
root.left.right = createNode('E');
root.right.left = createNode('F');

console.log("D 삭제 전:");
printHorizontal(root);

// D 삭제 실행
deleteLeaf(root, 'D');

console.log("\nD 삭제 후:");
printHorizontal(root);
```

깊이 우선 탐색(DFS)

```python
%%writefile search.js

function createNode(value) {
  return {
    value: value,
    left: null,
    right: null
  };
}

const root = createNode('A');

root.left = createNode('B');
root.right = createNode('C');

root.left.left = createNode('D');
root.left.right = createNode('E');

function printHorizontal(node) {
  console.log("--- Tree Structure ---");
  console.log("        " + node.value);
  console.log("      /   \\");
  console.log("     " + node.left.value + "     " + node.right.value);
  console.log("    / \\");
  console.log("   " + node.left.left.value + "   " + node.left.right.value);
}

function getInorder(node, result = []) {
  if (!node) return result;
  getInorder(node.left, result);
  result.push(node.value);
  getInorder(node.right, result);
  return result;
}

// 특정 노드를 찾는 함수 추가
function findNode(node, target) {
  if (!node) return null;
  if (node.value === target) return node;

  // 왼쪽 서브트리 검색
  let found = findNode(node.left, target);
  if (found) return found;

  // 오른쪽 서브트리 검색
  return findNode(node.right, target);
}

printHorizontal(root);
console.log("\n--- Inorder Traversal ---");
const traversalResult = getInorder(root);
console.log(traversalResult.join(" -> "));

// 검색 예시
const targetValue = 'E';
const foundNode = findNode(root, targetValue);
console.log("\n--- Search Result ---");
if (foundNode) {
  console.log(`Node with value '${targetValue}' found.`);
} else {
  console.log(`Node with value '${targetValue}' not found.`);
}
```

너비 우선 탐색(BFS)

```python
%%writefile search.js

function createNode(value) {
  return {
    value: value,
    left: null,
    right: null
  };
}

const root = createNode('A');

root.left = createNode('B');
root.right = createNode('C');

root.left.left = createNode('D');
root.left.right = createNode('E');

function printHorizontal(node) {
  console.log("--- Tree Structure ---");
  console.log("        " + node.value);
  console.log("      /   \\");
  console.log("     " + node.left.value + "     " + node.right.value);
  console.log("    / \\");
  console.log("   " + node.left.left.value + "   " + node.left.right.value);
}

// BFS(너비 우선 탐색)로 특정 노드를 찾는 함수
function findNodeBFS(root, target) {
  if (!root) return null;

  const queue = [root]; // 탐색을 위한 큐
  const visited = [];   // 방문한 노드 순서 저장

  while (queue.length > 0) {
    let node = queue.shift();
    visited.push(node.value);

    if (node.value === target) {
      console.log("Visit Path: " + visited.join(" -> "));
      return node;
    }

    if (node.left) queue.push(node.left);
    if (node.right) queue.push(node.right);
  }

  console.log("Visit Path: " + visited.join(" -> "));
  return null;
}

printHorizontal(root);

// BFS 검색 예시
const targetValue = 'E';
console.log(`\n--- BFS Search Result for '${targetValue}' ---`);
const foundNode = findNodeBFS(root, targetValue);

if (foundNode) {
  console.log(`Node with value '${targetValue}' found.`);
} else {
  console.log(`Node with value '${targetValue}' not found.`);
}
```

```python
%%writefile sql.js

// --- Base Structure ---
function createNode(value) {
  return {
    value: value,
    left: null,
    right: null
  };
}

let root = createNode('김철수(서울)');
root.left = createNode('이영희(부산)');
root.right = createNode('박민수(광주)');
root.left.left = createNode('최지우(제주)');
root.left.right = createNode('정해인(대구)');

// --- Directory Style Tree Visualization ---
function printDirectoryTree(node, prefix = "", isLeft = true) {
  if (!node) return;

  if (node.right) {
    printDirectoryTree(node.right, prefix + (isLeft ? "│   " : "    "), false);
  }

  console.log(prefix + (isLeft ? "└── " : "┌── ") + node.value);

  if (node.left) {
    printDirectoryTree(node.left, prefix + (isLeft ? "    " : "│   "), true);
  }
}

// --- SQL Engine Emulation ---

function selectNode(target) {
  if (!root) return null;
  const queue = [root];
  while (queue.length > 0) {
    let node = queue.shift();
    if (node.value === target) return node;
    if (node.left) queue.push(node.left);
    if (node.right) queue.push(node.right);
  }
  return null;
}

function executeSQL(action, params) {
  console.log(`\n>> EXECUTING SQL: ${action} ${JSON.stringify(params)}`);

  // Before state
  if (action !== 'SELECT') {
    console.log(`[${action} BEFORE]`);
    printDirectoryTree(root);
  }

  switch (action) {
    case 'SELECT':
      console.log(`[SQL Log] SELECT * FROM address_book WHERE name_addr = '${params.target}';`);
      const found = selectNode(params.target);
      console.log(found ? `   - Result: Found ${found.value}` : "   - Result: Not Found");
      break;

    case 'INSERT':
      console.log(`[SQL Log] INSERT INTO address_book (name_addr) VALUES ('${params.value}');`);
      const newNode = createNode(params.value);
      const queue = [root];
      while (queue.length > 0) {
        let node = queue.shift();
        if (!node.left) { node.left = newNode; break; }
        else queue.push(node.left);
        if (!node.right) { node.right = newNode; break; }
        else queue.push(node.right);
      }
      break;

    case 'UPDATE':
      console.log(`[SQL Log] UPDATE address_book SET name_addr = '${params.newValue}' WHERE name_addr = '${params.oldValue}';`);
      const targetNode = selectNode(params.oldValue);
      if (targetNode) targetNode.value = params.newValue;
      break;

    case 'DELETE':
      console.log(`[SQL Log] DELETE FROM address_book WHERE name_addr = '${params.target}';`);
      const dQueue = [root];
      while (dQueue.length > 0) {
        let node = dQueue.shift();
        if (node.left && node.left.value === params.target) { node.left = null; break; }
        if (node.right && node.right.value === params.target) { node.right = null; break; }
        if (node.left) dQueue.push(node.left);
        if (node.right) dQueue.push(node.right);
      }
      break;
  }

  // After state
  if (action !== 'SELECT') {
    console.log(`[${action} AFTER]`);
    printDirectoryTree(root);
  }
  console.log("--------------------------------------------------");
}

// --- Scenario Execution ---
console.log("===== MariaDB Emulation Start =====");

executeSQL('SELECT', { target: '최지우(제주)' });
executeSQL('UPDATE', { oldValue: '박민수(광주)', newValue: '박민수(인천)' });
executeSQL('DELETE', { target: '최지우(제주)' });
executeSQL('INSERT', { value: '홍길동(강원)' });

console.log("===== MariaDB Emulation End =====");
```

```python
%%writefile server.js

const express = require('express');
const path = require('path');
const mysql = require('mysql2');

const app = express();
const PORT = 3000;

app.use(express.urlencoded({ extended: true }));
app.use(express.json()); // JSON body 처리

const conn = mysql.createConnection({
    host: 'localhost',
    user: 'testuser',
    password: '1234',
    database: 'testdb'
});

conn.connect((err) => {
    if (err) {
        console.error('DB 연결 실패:', err);
        return;
    }
    console.log('DB 연결 성공');
});

// --- 메인 페이지 ---
app.get('/', (req, res) => {
    res.sendFile(path.join(__dirname, 'index.html'));
});

// --- 트리 구조 ---
let root = null;

// Node 생성
function createNode(id, value) {
    return {
        id,
        value,
        left: null,
        right: null
    };
}

// DB → 트리 변환
function buildTree(rows) {
    const map = {};
    rows.forEach(row => {
        map[row.id] = createNode(row.id, row.name_addr);
    });

    let rootNode = null;
    rows.forEach(row => {
        if (!rootNode) rootNode = map[row.id];
        if (row.left_child_id) map[row.id].left = map[row.left_child_id];
        if (row.right_child_id) map[row.id].right = map[row.right_child_id];
    });

    return rootNode;
}

// 트리 초기 로딩
function loadTree() {
    return new Promise((resolve, reject) => {
        conn.query("SELECT * FROM address_book", (err, rows) => {
            if (err) return reject(err);
            root = buildTree(rows);
            resolve();
        });
    });
}

// --- BFS 검색 ---
function selectNode(target) {
    if (!root) return null;
    const queue = [root];
    while (queue.length > 0) {
        const node = queue.shift();
        if (node.value === target) return node;
        if (node.left) queue.push(node.left);
        if (node.right) queue.push(node.right);
    }
    return null;
}

// --- INSERT ---
async function insertNode(value) {
    return new Promise((resolve, reject) => {
        conn.query(
            "INSERT INTO address_book (name_addr) VALUES (?)",
            [value],
            (err, result) => {
                if (err) return reject(err);

                const newNode = createNode(result.insertId, value);

                if (!root) {
                    root = newNode;
                    return resolve();
                }

                const queue = [root];
                while (queue.length > 0) {
                    let node = queue.shift();

                    if (!node.left) {
                        node.left = newNode;
                        conn.query(
                            "UPDATE address_book SET left_child_id=? WHERE id=?",
                            [newNode.id, node.id]
                        );
                        break;
                    } else queue.push(node.left);

                    if (!node.right) {
                        node.right = newNode;
                        conn.query(
                            "UPDATE address_book SET right_child_id=? WHERE id=?",
                            [newNode.id, node.id]
                        );
                        break;
                    } else queue.push(node.right);
                }

                resolve();
            }
        );
    });
}

// --- UPDATE ---
function updateNode(oldValue, newValue) {
    return new Promise((resolve, reject) => {
        conn.query(
            "UPDATE address_book SET name_addr=? WHERE name_addr=?",
            [newValue, oldValue],
            (err) => {
                if (err) return reject(err);

                const node = selectNode(oldValue);
                if (node) node.value = newValue;

                resolve();
            }
        );
    });
}

// --- DELETE ---
function deleteNode(target) {
    return new Promise((resolve, reject) => {
        conn.query(
            "DELETE FROM address_book WHERE name_addr=?",
            [target],
            (err) => {
                if (err) return reject(err);

                const queue = [root];
                while (queue.length > 0) {
                    const node = queue.shift();

                    if (node.left && node.left.value === target) {
                        node.left = null;
                        break;
                    }
                    if (node.right && node.right.value === target) {
                        node.right = null;
                        break;
                    }

                    if (node.left) queue.push(node.left);
                    if (node.right) queue.push(node.right);
                }

                resolve();
            }
        );
    });
}

// --- 트리 문자열화 ---
function printDirectoryTreeStr(node, prefix = "", isLeft = true) {
    if (!node) return "";

    let str = "";
    if (node.right) {
        str += printDirectoryTreeStr(node.right, prefix + (isLeft ? "│   " : "    "), false);
    }

    str += prefix + (isLeft ? "└── " : "┌── ") + node.value + "\n";

    if (node.left) {
        str += printDirectoryTreeStr(node.left, prefix + (isLeft ? "    " : "│   "), true);
    }

    return str;
}

function getTreeString() {
    return root ? printDirectoryTreeStr(root) : "(트리가 비어있습니다)";
}

// --- API ---
app.get('/run', async (req, res) => {
    try {
        await loadTree();
        res.send(getTreeString());
    } catch (err) {
        res.status(500).json({ error: err.toString() });
    }
});

app.post('/insert', async (req, res) => {
    try {
        await insertNode(req.body.value || '새 노드');
        res.send(getTreeString());
    } catch (err) {
        res.status(500).json({ error: err.toString() });
    }
});

app.post('/update', async (req, res) => {
    try {
        await updateNode(req.body.oldValue, req.body.newValue);
        res.send(getTreeString());
    } catch (err) {
        res.status(500).json({ error: err.toString() });
    }
});

app.post('/delete', async (req, res) => {
    try {
        await deleteNode(req.body.target);
        res.send(getTreeString());
    } catch (err) {
        res.status(500).json({ error: err.toString() });
    }
});

app.listen(PORT, () => {
    console.log(`서버 실행 중: ${PORT}`);
});
```

```python
function selectNode(target) {
  if (!root) return null;
  const queue = [root];
  while (queue.length > 0) {
    let node = queue.shift(); // (A) 큐에서 노드 추출
    if (node.value === target) return node;
    if (node.left) queue.push(node.left); // (B) 왼쪽 자식 큐에 추가
    if (node.right) queue.push(node.right); // (C) 오른쪽 자식 큐에 추가
  }
  return null;
}
```
