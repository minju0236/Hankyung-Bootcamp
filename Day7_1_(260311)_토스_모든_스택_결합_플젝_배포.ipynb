```python
%%writefile server.js

const express = require('express');
const path = require('path');
const app = express();
const PORT = 3000;

// [1] ž†ēž†Ā žěźžõź(JS ŽćįžĚīŪĄį) Í≤ĹŽ°ú žĄ§ž†ē
app.use('/static', express.static(path.join(__dirname, 'static')));

// [2] ŪéėžĚīžßÄŽ≥Ą ÍįúŽ≥Ą ŽĚľžöįŪäł (Route) žĄ§ž†ē
app.get('/', (req, res) => {
    res.sendFile(path.join(__dirname, 'templates', 'home.html'));
});

// HTMLžóźžĄú script.js Ž°ú Ž∂ąŽü¨žė§ŽäĒ Í≤Ĺžöį ŽĆÄžĚĎ
app.use(express.static(path.join(__dirname, 'static')));

app.get('/home', (req, res) => {
    res.sendFile(path.join(__dirname, 'templates', 'home.html'));
});

app.get('/account', (req, res) => {
    res.sendFile(path.join(__dirname, 'templates', 'account.html'));
});

app.get('/finance', (req, res) => {
    res.sendFile(path.join(__dirname, 'templates', 'finance.html'));
});

app.get('/transfer', (req, res) => {
    res.sendFile(path.join(__dirname, 'templates', 'transfer.html'));
});

app.get('/customer', (req, res) => {
    res.sendFile(path.join(__dirname, 'templates', 'customer.html'));
});

app.get('/atm', (req, res) => {
    res.sendFile(path.join(__dirname, 'templates', 'atm.html'));
});

app.listen(PORT, () => {
    console.log(`=========================================`);
    console.log(` ÍłąžúĶ žĄúŽĻĄžä§ žĄúŽ≤ĄÍįÄ ŪŹ¨Ūäł ${PORT}žóźžĄú žěĎŽŹô ž§ĎžěÖŽčąŽč§.`);
    console.log(`=========================================`);
});
```
