화살표 함수에서 this는 상위스코프를 의미해, 여기서는 userProfile이므로 this.name은 박지민을 의미함.

일반 함수의 this는 상위스코프가 아닌 window를 의미해, this.name으로 접근 시 undefined이 출력된다. 범위의 문제에서는 에러출력이 아니라서 문법 오류로 처리하지 않는다. 그래서 undefined로 출력된다.

```python
%%writefile lexical.js

const userProfile = {
    name: "박지민",
    friends: ["최현우", "임윤아"],

    printFriends() {
        // 일반 함수의 콜백으로 화살표 함수 사용
        this.friends.forEach((friend) => {
            // 화살표 함수는 상위 스코프인 printFriends의 this(userProfile)를 참조함
            console.log(`${this.name}의 친구: ${friend}`);
        });
    }
};

userProfile.printFriends();
```

```python
!node lexical.js
```

```python
%%writefile this.js

const userProfile = {
    name: "박지민",
    friends: ["최현우", "임윤아"],

    printFriends() {
        // forEach의 콜백으로 '일반 함수'를 사용
        this.friends.forEach(function(friend) {
            // 이 함수는 객체의 메서드로 호출된 게 아니라, 내부의 '일반 함수'로 실행됨
            // 따라서 여기서의 this는 userProfile이 아니라 전역 객체(window/undefined)를 가리킴
            console.log(`${this.name}의 친구: ${friend}`);
        });
    }
};

userProfile.printFriends();
// 출력 결과: "undefined의 친구: 최현우", "undefined의 친구: 임윤아" (엄격 모드 기준)
```

```python
!node this.js
```

```python
%%writefile this.js

const userProfile = {
    name: "박지민",
    friends: ["최현우", "임윤아"],

    printFriends() {
        const self = this; // this(userProfile)를 변수에 저장
        this.friends.forEach(function(friend) {
            console.log(`${self.name}의 친구: ${friend}`); // self를 통해 접근
        });
    }
};

userProfile.printFriends();
```

```python
!node this.js
```
