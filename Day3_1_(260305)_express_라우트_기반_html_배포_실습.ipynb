```python
%%writefile server.js

// server.js
// - Express로 "라우트 기반" 정적 HTML 사이트를 제공하는 예제
// - 목표: /, /about, /products 같은 "가상 URL 경로(route)"로만 페이지 접근
// - 파일들은 모두 Colab의 /content/ (즉 __dirname) 폴더에 있다고 가정

// 1) Express 프레임워크 로드
//    - Node.js에서 HTTP 서버를 쉽게 구성하게 해주는 웹 프레임워크
const express = require("express");

// 2) path 모듈 로드
//    - 파일 경로를 OS에 맞게 안전하게 조합해주는 유틸리티
//    - (Windows는 \, Linux/Mac은 / 이므로 join을 쓰는 습관이 중요)
const path = require("path");

// 3) Express 애플리케이션(서버) 인스턴스 생성
//    - app 객체에 라우트(경로)와 핸들러(동작)를 등록한다
const app = express();

/* ============================================================
   HTML 라우트
   - "URL 경로"와 "파일 경로"는 완전히 다름
   - 아래 코드는 URL을 파일로 매핑하는 규칙을 수동으로 만든 것
   - express.static()을 사용하지 않으므로
     /index.html 같은 "파일명 직접 접근"은 기본적으로 제공되지 않음
   ============================================================ */

// [1] Home 라우트
// - 사용자가 브라우저에서 아래 URL로 접근하면 실행됨:
//   http://localhost:3000/
// - 라우트 "/"는 웹사이트의 루트 URL(가상 경로)이지, 파일 시스템의 루트(/)가 아님
app.get("/", (req, res) => {
  // res.sendFile(절대경로)
  // - 서버가 해당 파일(HTML)을 그대로 읽어서 HTTP 응답으로 전송
  // __dirname: 현재 server.js 파일이 위치한 폴더의 절대 경로
  // Colab에서는 보통 /content 가 됨
  res.sendFile(path.join(__dirname, "index.html"));
});

// [2] 소개 페이지 라우트
// - URL: http://localhost:3000/about
// - 실제 파일: /content/about.html
app.get("/about", (req, res) => {
  res.sendFile(path.join(__dirname, "about.html"));
});

// [3] 상품 페이지 라우트
// - URL: http://localhost:3000/products
// - 실제 파일: /content/products.html
app.get("/products", (req, res) => {
  res.sendFile(path.join(__dirname, "products.html"));
});

// [4] 대출 페이지 라우트
// - URL: http://localhost:3000/loan
// - 실제 파일: /content/loan.html
app.get("/loan", (req, res) => {
  res.sendFile(path.join(__dirname, "loan.html"));
});

// [5] 고객센터 페이지 라우트
// - URL: http://localhost:3000/support
// - 실제 파일: /content/support.html
app.get("/support", (req, res) => {
  res.sendFile(path.join(__dirname, "support.html"));
});

/* ============================================================
   로고 이미지 라우트
   - 현재 구성은 express.static()을 사용하지 않기 때문에
     이미지도 "라우트"로 직접 제공해야 브라우저에서 로드됨
   - index.html에서만 로고를 쓴다고 했으므로 1개 라우트만 추가
   - index.html에서는 아래처럼 작성되어야 함:
     <img src="/logo.png" ...>
   ============================================================ */

// [6] 로고 이미지 요청 라우트
// - URL: http://localhost:3000/logo.png
// - 실제 파일: /content/logo.png
app.get("/img/:file", (req, res) => {
  res.sendFile(path.join(__dirname, "img", req.params.file));
});

/* ============================================================
   서버 실행
   - 포트 3000에서 HTTP 서버를 시작
   - Colab에서는 localhost로 직접 접근이 어려울 수 있어
     Cloudflare Tunnel/Ngrok 등을 통해 외부에서 접속하는 경우가 많음
   ============================================================ */

app.listen(3000, () => {
  // 서버가 시작되면 터미널(Colab 출력)에 안내 메시지 출력
  console.log("Server running on port 3000");
});
```
