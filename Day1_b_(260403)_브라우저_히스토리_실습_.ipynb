```python
%%writefile test4.js

const backStack = [];
const forwardStack = [];

// 페이지 이동
function visit(url) {
    backStack.push(url);
    forwardStack.length = 0; // 앞으로가기 초기화
}

function goBack() {
    if (backStack.length <= 1) return;

    const current = backStack.pop();
    forwardStack.push(current);

    return backStack[backStack.length - 1];
}

function goForward() {
    const page = forwardStack.pop();
    backStack.push(page);
    return page;
}

// 실행
visit('https://www.google.com');
visit('https://www.naver.com');
visit('https://www.youtube.com');

goBack();        // 1
visit('https://www.github.com'); // 2
goBack();        // 3
goForward();     // 4
goForward();     // 5

// 실행
console.log('backStack:', backStack);
console.log('forwardStack:', forwardStack);
```

```python
%%writefile test5.js

const backStack = [];
const forwardStack = [];

// 페이지 이동
function visit(url) {
    backStack.push(url);
    forwardStack.length = 0; // 앞으로가기 초기화

    if (backStack.length > 3) { // backStack의 길이가 3을 넘으면
      backStack.shift(); // 앞에서 하나 제거
    }
}

function goBack() {
    if (backStack.length <= 1) return;

    const current = backStack.pop();
    forwardStack.push(current);

    return backStack[backStack.length - 1];
}

function goForward() {
    const page = forwardStack.pop();
    backStack.push(page);
    return page;
}

// 실행
visit('https://www.google.com');
visit('https://www.naver.com');
visit('https://www.youtube.com');
visit('https://www.toss.com');
visit('https://www.kakao.com');

// 실행
console.log('backStack:', backStack);
console.log('forwardStack:', forwardStack);
```

```python
%%writefile user.js

// 추천 시스템 예제 4: 현실 흐름

const userGraph = {
    '사용자A': ['사용자B', '사용자C'],
    '사용자B': ['사용자A', '사용자D'],
    '사용자C': ['사용자A', '사용자E'],
    '사용자D': ['사용자B'],
    '사용자E': ['사용자C']
};

function recommendUsers(user) {
    const direct = userGraph[user] || [];
    const result = new Set();

    for (const friend of direct) {
        const linkedUsers = userGraph[friend] || [];

        for (const candidate of linkedUsers) {
            if (candidate !== user && !direct.includes(candidate)) {
                result.add(candidate);
            }
        }
    }

    return [...result];
}

console.log(recommendUsers('사용자A'));
// ['사용자D', '사용자E']
```

```python
%%writefile user.js

// 추천 시스템 예제 4: 현실 흐름

const userGraph = {
    '김민수': ['박지훈', '이수빈'],
    '박지훈': ['김민수', '최유진'],
    '이수빈': ['김민수', '정하늘'],
    '최유진': ['박지훈'],
    '정하늘': ['이수빈']
};

function recommendFriends(user) {
    const direct = userGraph[user] || [];
    const result = new Set();

    for (const friend of direct) {
        const linkedUsers = userGraph[friend] || [];

        for (const candidate of linkedUsers) {
            if (candidate !== user && !direct.includes(candidate)) {
                result.add(candidate);
            }
        }
    }

    return [...result];
}

// 추천 시스템 예제 5: 상태 확인

function printRecommendation(user) {
    console.log('기준 사용자:', user);
    console.log('추천 결과:', recommendFriends(user));
}

// 실행
printRecommendation('김민수');
```

```python
%%writefile lru.js

// 캐시 시스템 예제 1: LRU 캐시 클래스

class LRUCache {
    constructor(limit) {
        this.limit = limit;
        this.cache = new Map();
    }

    get(key) {
        if (!this.cache.has(key)) return null;

        const value = this.cache.get(key);

        this.cache.delete(key);
        this.cache.set(key, value);

        return value;
    }

    put(key, value) {
        if (this.cache.has(key)) {
            this.cache.delete(key);
        }

        this.cache.set(key, value);

        if (this.cache.size > this.limit) {
            const oldestKey = this.cache.keys().next().value;
            this.cache.delete(oldestKey);
        }
    }
}

const productCache = new LRUCache(3);

console.log('LRU 캐시 생성 완료');

// 캐시 시스템 예제 2: 캐시 조회

productCache.put('product:101', '노트북');
productCache.put('product:102', '키보드');
productCache.put('product:103', '마우스');

console.log(productCache.get('product:102')); // 키보드
console.log(productCache.get('product:999')); // null

// 캐시 시스템 예제 3: 오래된 데이터 제거

productCache.put('product:101', '노트북');
productCache.put('product:102', '키보드');
productCache.put('product:103', '마우스');

productCache.get('product:101'); // 최근 사용 갱신
productCache.put('product:104', '모니터'); // 용량 초과 → 가장 오래된 항목 제거

console.log([...productCache.cache.entries()]);

// 결과
// [
//   ['product:103', '마우스'],
//   ['product:101', '노트북'],
//   ['product:104', '모니터']
// ]
```
