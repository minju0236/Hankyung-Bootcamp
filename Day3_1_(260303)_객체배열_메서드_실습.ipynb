1. name과 price라는 속성을 가진 items이라는 객체들의 배열을 정의 (name은 string, price는 number 타입)

2. items 중 price가 1000 이상인 객체에 대하여 필터링 후 toUpperCase 메서드를 이용하여 name을 대문자로 변환하고 최종적으로 result라는 변수에 담음 (원본인 items는 변화X)

3. result 출력: price가 1000원 이상인 객체의 name을 대문자로 출력



```python
%%writefile item.js

const items = [
  { name: 'apple', price: 1200 },
  { name: 'pear', price: 800 },
  { name: 'orange', price: 1500 }
];

const result = items
  .filter(item => item.price >= 1000) // 1. 필터링
  .map(item => item.name.toUpperCase()); // 2. 이름 변환

console.log(result); // ['APPLE', 'ORANGE']
```

장바구니 총 결제 금액 계산

1. 먼저 cart라는 객체배열을 정의한다. 각각의 객체는 이름과 가격을 속성으로 가진다.

2. totalPrice라는 변수를 정의하고, cart에서 reduce 메서드를 사용한다. 초깃값은 0이고, cur.price로 가격을 하나씩 받아오며 acc에는 이전의 값을 저장한다. 즉 cart 내부의 모든 가격을 더해서 totalPrice에 저장한다.

3. 최종적으로 총 결제 금액인 totalPrice을 `총 결제 금액: ${totalPrice}원` 문자열로 출력한다.

```python
%%writefile reduce.js

const cart = [
  { name: '셔츠', price: 35000 },
  { name: '바지', price: 48000 },
  { name: '양말', price: 5000 }
];

// TODO: reduce를 사용하여 총 합계를 구하세요.
const totalPrice = cart.reduce((acc, cur) => acc + cur.price, 0);

/* reduce는 값의 축약을 의미
const totalPrice = cart.reduce((acc, cur) => acc + cur.price, 0); */
//                             ───  ───      ─────────────    ─
//                              ▲    ▲             ▲          ▲
//                              │    │             │          │
// [1] acc (Accumulator) ───────┘    │             │          │
//     : '누적 합계'가 저장되는 바구니 (이전 단계의 결과값)   │          │
//                                   │             │          │
// [2] cur (Current Value) ──────────┘             │          │
//     : '현재 요소', 즉 배열에서 지금 막 꺼낸 상품 객체     │          │
//                                                 │          │
// [3] 실행부 (acc + cur.price)   ───────────────────┘          │
//     : 지금까지의 합계(acc)에 현재 상품 가격(price)을 더함            │
//                                                            │
// [4] 0 (Initial Value)   ───────────────────────────────────┘
//     : '시작 숫자', 계산기를 0으로 리셋하고 시작한다는 의미


console.log(`총 결제 금액: ${totalPrice}원`);
```

```python
!node reduce.js
```

고득점자 명단 가공하기

1. 먼저 scores라는 객체배열을 정의한다. 각각의 객체는 이름과 가격을 점수를 속성으로 가진다.

2. honorsList라는 변수를 정의한다. scores에서 점수가 80이상인 객체들만 필터링하고, map으로 객체의 name속성을 반환하며, name 속성값인 이름을 오름차순으로 정렬해 honorsList에 저장한다.

3. 최종적으로 honorsList에는 점수가 80점 이상인 학생들의 이름이 출력된다.

```python
%%writefile score.js

const scores = [
  { name: '김철수', score: 70 },
  { name: '이영희', score: 95 },
  { name: '박지성', score: 88 },
  { name: '최유리', score: 60 },
  { name: '정우성', score: 92 }
];

// TODO: 80점 이상 -> 이름만 추출 -> 가나다순 정렬
const honorsList = scores
  .filter(s => s.score >= 80)
  .map(s => s.name)
  .sort();

console.log(honorsList); // ["박지성", "이영희", "정우성"]
```
