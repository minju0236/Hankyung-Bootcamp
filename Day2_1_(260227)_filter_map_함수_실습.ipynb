```python
%%writefile callback.js
/**
 * 콜백 함수의 유연성 실습
 * 목적: 메인 로직(순회)은 유지하되, 조건(판단)만 교체하여 함수를 무한히 재활용함
 */

// 1. 메인 엔진 (필터 틀)
// arr: 데이터 배열, condition: 참/거짓을 판단할 콜백 함수
const filterArr = (arr, condition) => {
    const result = [];
    for (const item of arr) {
        // 콜백 함수(condition)에 현재 요소를 전달하여 결과가 true일 때만 담음
        if (condition(item)) {
            result.push(item);
        }
    }
    return result;
};

// 2. 테스트 데이터
const numbers = [1, 5, 8, 10, 12, 15, 20, 27];

// 3. 콜백 함수만 바꿔가며 호출하기
console.log("--- 콜백 함수의 유연한 변신 ---");

// (1) 홀수 찾기: n을 2로 나눈 나머지가 있으면 홀수
// condition은 n => n % 2 !== 0; 부분
const odds = filterArr(numbers, n => n % 2 !== 0);
console.log("1. 홀수 리스트:", odds);

// (2) 짝수 찾기: n을 2로 나눈 나머지가 없으면 짝수
// condition은 n => n % 2 === 0; 부분
const evens = filterArr(numbers, n => n % 2 === 0);
console.log("2. 짝수 리스트:", evens);

// (3) 10보다 큰 수 찾기
// condition은 n => n > 10; 부분
const biggerThanTen = filterArr(numbers, n => n > 10);
console.log("3. 10보다 큰 수:", biggerThanTen);

// (4) 5의 배수 찾기
// condition은 n => n % 5 === 0; 부분
const multiplesOfFive = filterArr(numbers, n => n % 5 === 0);
console.log("4. 5의 배수:", multiplesOfFive);

// (5) 특정 범위 (10 이상 20 이하) 찾기
// condition n >= 10 && n <= 20; 부분
const rangeValues = filterArr(numbers, n => n >= 10 && n <= 20);
console.log("5. 10~20 사이 값:", rangeValues);

// (6) 응용: 제곱했을 때 100이 넘는 수 찾기
// condition n * n > 100; 부분
const powerLarge = filterArr(numbers, n => n * n > 100);
console.log("6. 제곱이 100 초과:", powerLarge);
```

```python
%%writefile async.js
/**
 * 자바스크립트 비동기 핵심: "멈추지 않는 엔진"
 * 1. 서버 응답을 기다리는 동안(setTimeout)
 * 2. 메인 스레드는 다음 코드를 즉시 실행하고
 * 3. 작업이 완료되면 '태스크 큐'에 담겨있던 콜백이 실행됨
 */

// [함수 정의] 서버에서 데이터를 가져오는 함수
const getDataFromServer = (id, callback) => {
    console.log(`\n[STEP 1] 서버에 ${id}번 데이터 요청을 보냈습니다.`);

    // 서버 통신 시뮬레이션 (3초 대기)
    // setTimeout(,3000) -> 3초 후 실행
    setTimeout(() => {
        const dummyData = {
            userId: id,
            title: "자바스크립트 비동기 마스터",
            content: "콜백 함수는 비동기 처리의 기초입니다."
        };
        console.log(`\n[STEP 3] 서버 응답이 도착했습니다. (3초 경과)`);

        // 데이터가 준비되었을 때만 콜백 실행
        // renderPage에 파라미터로 dummyData를 넣음
        callback(dummyData);
    }, 3000);
};

// [함수 정의] 데이터를 받아서 화면에 그리는 함수 (콜백용)
const renderPage = (data) => {
    console.log(`[STEP 4] 페이지 렌더링: ${data.title}`);
    console.log(`         내용: ${data.content}`);
    console.log("========================================\n");
};

// --- 실제 실행 흐름 ---

console.log(">>> 프로그램을 시작합니다.");

// 비동기 함수 호출 (renderPage 함수를 '나중에 실행할 일'로 전달)
// getDataFromServer에 renderPage를 파라미터로 넣음
getDataFromServer(5, renderPage);

// 중요: 아래 코드는 STEP 1 직후에 바로 실행됩니다. (기다리지 않음)
console.log(">>> [STEP 2] 다른 로직 실행 중...");
console.log("    - 배너 광고 로딩 중...");
console.log("    - 실시간 채팅 연결 중...");

let seconds = 1;
// setInterval: ~초마다 실행 (여기서는 1000이므로 1초마다 실행)
const timer = setInterval(() => {
    if (seconds >= 3) {
        clearInterval(timer); // timer interval 끝내기 - 3 이상이므로 2초까지만 실행
    } else {
        console.log(`    ... 서버 기다리는 중 (${seconds++}초)`); // 1씩 증가
    }
}, 1000);
```

filterScores라는 함수를 선언하고 scores라는 배열과 (score) => score >= 60함수를 predicate라는 변수로 받아 해당 로직 리턴값을 passed로 반환하고 출력

```python
%%writefile test.js

function filterScores(scores, predicate) {
  const passed = [];
  for (let i = 0; i < scores.length; i++) {
    if (predicate(scores[i])) passed.push(scores[i]);
  }
  return passed;
}

const scores = [45, 72, 88, 59, 60];

const pass = filterScores(scores, (score) => score >= 60);
console.log(pass);
```

1. 먼저 filter함수와 map함수 선언하였다.

2. filter함수는 predicate함수를 돌아서 true면 out이라는 빈 배열에 arr[i]를 추가하여 리턴한다.

3. map함수는 transform함수를 돌아서 out이라는 빈 배열에 transform함수를 돌아서 나온 값을 추가한 후 out을 리턴한다.

4. sum은 total에 파라미터로 받은 arr의 값들을 모두 더해서 최종 리턴한다.

5. prices는 고정 배열(회원들의 구매 가격)

6. filtered는 filter함수에 (p) => p >= 10000 로직을 predicate콜백함수로 지정하여 만원 이상 구매한 회원들만 배열에 저장한다.

7. discounted는 Math.floor(p * 0.9)을 통해 값을 90%로 만드는, 즉 10%를 할인 적용하도록 한다. map을 통해 앞서 만원 이상 구매한 회원들만 저장한 filtered와, 10% 할인을 적용하는 로직을 인자로 받아 최종적으로 만원이상 구매한 회원들에게 10%할인을 적용한 값들의 배열을 리턴한다.

8. 마지막으로 total에서는 discounted의 배열을 모두 더하여 최종 합계를 저장한다.




```python
%%writefile test.js

function filter(arr, predicate) {
  const out = [];
  for (let i = 0; i < arr.length; i++) {
    if (predicate(arr[i])) out.push(arr[i]);
  }
  return out;
}

function map(arr, transform) {
  const out = [];
  for (let i = 0; i < arr.length; i++) {
    out.push(transform(arr[i]));
  }
  return out;
}

function sum(arr) {
  let total = 0;
  for (let i = 0; i < arr.length; i++) total += arr[i];
  return total;
}

// 데이터
const prices = [12000, 8000, 30000, 15000, 4000];

// 1) 10000원 이상만 구매(조건 콜백)
const filtered = filter(prices, (p) => p >= 10000);

// 2) 10% 할인 적용(변환 콜백)
const discounted = map(filtered, (p) => Math.floor(p * 0.9));

// 3) 합계
const total = sum(discounted);

console.log("원가:", prices);
console.log("필터 후:", filtered);
console.log("할인 후:", discounted);
console.log("최종 합계:", total);
```
