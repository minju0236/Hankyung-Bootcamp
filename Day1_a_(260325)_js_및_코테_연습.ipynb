```python
%%writefile stack.js

// 스택 예제 1: 기본 Push & Pop
const stack = [];

// 1. 데이터 삽입 (Push)
stack.push('데이터 A');
stack.push('데이터 B');
stack.push('데이터 C');
// 현재 상태: ['데이터 A', '데이터 B', '데이터 C'] (Top: 데이터 C)

// 2. 데이터 추출 (Pop)
const popped = stack.pop();

console.log(`추출된 요소: ${popped}`); // 추출된 요소: 데이터 C
console.log(`현재 스택 상태: ${stack}`); // ['데이터 A', '데이터 B'] 이 형식을 기억하고 명령대기 출력없음
```

```python
%%writefile reverse.js

// 스택 예제 3: 문자열 뒤집기
const text = "HELLO";
const charStack = [];
let reversedText = "";

// 1. 글자를 하나씩 스택에 Push
for (let char of text) {
    charStack.push(char);
}

// 2. 스택에서 하나씩 Pop하여 새로운 문자열 생성
while (charStack.length > 0) {
    reversedText += charStack.pop();
}

console.log(`역순 결과: ${reversedText}`); // 역순 결과: OLLEH
console.log(`스택 상태: ${charStack}`); // [] 이 형식을 기억하고 명령대기 출력없음
```

```python
%%writefile parentheses.js

// 스택 예제 4: 괄호 짝 맞추기
const expression = "( ( 1 + 2 ) * 3 )";
const parenStack = [];
let isBalanced = true;

for (let char of expression) {
    if (char === '(') {
        parenStack.push(char); // 여는 괄호는 무조건 Push
    } else if (char === ')') {
        if (parenStack.length === 0) {
            isBalanced = false;
            break;
        }
        parenStack.pop(); // 짝이 맞으면 최상단 여는 괄호 제거
    }
}

if (parenStack.length !== 0) isBalanced = false;

console.log(`수식의 유효성: ${isBalanced}`); // 수식의 유효성: true
console.log(`최종 스택: ${parenStack}`); // [] 이 형식을 기억하고 명령대기 출력없음
```

```python
!node parentheses.js
```

```python
%%writefile queue.js

// 큐 예제 1: 기본 Enqueue & Dequeue
const queue = [];

// 1. 데이터 삽입 (Enqueue)
queue.push('작업 1');
queue.push('작업 2');
queue.push('작업 3');
// 현재 상태: ['작업 1', '작업 2', '작업 3'] (Front: 작업 1)

// 2. 데이터 추출 (Dequeue)
const task = queue.shift();

console.log(`처리된 작업: ${task}`); // 처리된 작업: 작업 1
console.log(`현재 큐 상태: ${queue}`); // ['작업 2', '작업 3'] 이 형식을 기억하고 명령대기 출력없음
```

```python
%%writefile list.js

// 큐 예제 4: 객체를 활용한 효율적 큐 (O(1) 구현)
class SimpleQueue {
  constructor() {
    this.storage = {};
    this.front = 0;
    this.rear = 0;
  }

  enqueue(item) {
    this.storage[this.rear] = item;
    this.rear++;
  }

  dequeue() {
    const item = this.storage[this.front];
    delete this.storage[this.front];
    this.front++;
    return item;
  }
}

const fastQueue = new SimpleQueue();
fastQueue.enqueue('데이터 1');
fastQueue.enqueue('데이터 2');

console.log(`추출 데이터: ${fastQueue.dequeue()}`); // 추출 데이터: 데이터 1
// 내부 인덱스 기반으로 동작하여 시프팅 비용이 없음
```

```python
%%writefile callback.js

// 큐 예제 5: 비동기 작업 대기열
const taskQueue = [];

// 비동기 작업 완료 순서대로 큐에 담김
taskQueue.push(() => console.log("첫 번째 콜백 실행"));
taskQueue.push(() => console.log("두 번째 콜백 실행"));

// 이벤트 루프가 큐에서 하나씩 꺼내어 실행
while (taskQueue.length > 0) {
    const callback = taskQueue.shift();
    callback();
}

console.log(`모든 태스크 완료: ${taskQueue}`); // [] 이 형식을 기억하고 명령대기 출력없음
```

```python
%%writefile deque.js

// 덱 예제 1: 양방향 삽입과 삭제
const deque = ['Data2', 'Data3'];

// 1. 앞쪽에 추가 (Unshift) & 뒤쪽에 추가 (Push)
deque.unshift('Data1');
deque.push('Data4');
// 상태: ['Data1', 'Data2', 'Data3', 'Data4']

// 2. 앞쪽에서 제거 (Shift) & 뒤쪽에서 제거 (Pop)
const frontItem = deque.shift(); // 'Data1'
const rearItem = deque.pop();    // 'Data4'

console.log(`앞/뒤 제거 결과: ${frontItem}, ${rearItem}`);
console.log(`현재 덱 상태: ${deque}`); // ['Data2', 'Data3'] 이 형식을 기억하고 명령대기 출력없음
```

```python
%%writefile codingTest1.js

const input = "(())()";
const arr = input.split('');
const stack = [];

for (let i = 0; i < arr.length; i++) {
  const state = arr[i];
  if (state === "(") {
    stack.push(state);
  } else {
    if (stack.length === 0) {
      console.log(false);
      return;
    }
    stack.pop();
  }
}

if (stack.length === 0) {
  console.log(true);
} else {
  console.log(false);
}
```

```python
%%writefile codingTest2.js

const input = ['김', '이', '박', '최', '정'];

let name = '';
let count = 1;

while (true) {
  if (input.length === 0) break;

  name = input.shift();

  if (count % 3 == 0) {
    count = 1;
  } else {
    input.push(name);
    count++;
  }
}

console.log(`마지막 생존자: ${name}`);
```

```python
%%writefile codingTest3.js

const input = [10, 20, 10, 30, 20];
const size = 3;
let count = 0;
let maxSum = 0;

while (true) {
  if (count == input.length * size) break;

  let sum = 0;
  for (let i = 0; i < size; i++) {
    const num = input.shift();
    sum += num;
    input.push(num);
  }

  count++;
  if (sum > maxSum) maxSum = sum;
}

console.log(`최대 합계: ${maxSum}`);
```

```python
!node codingTest3.js
```

```python
%%writefile codingTest4.js

const input = [[2, '김'], [1, '이'], [3, '박'], [2, '최']];
const pq = input.map(item => item[0]).sort((a, b) => b - a);
const person = '김';

let count = 0;
while (true) {
  if (input.length == 0) break;

  const name = input.shift();

  if (name[0] === pq[0]) {
    count++;
    pq.shift();

    if (name[1] === person) {
      console.log(count);
      return;
    }

  } else {
    input.push(name);
  }
}
```
