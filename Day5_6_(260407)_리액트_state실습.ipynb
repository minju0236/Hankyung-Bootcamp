```python
%%writefile /content/bootcamp/src/App.jsx

import { useState, useEffect } from 'react';

function App() {
  const [count, setCount] = useState(0); // count 선언
  const [double, setDouble] = useState(0); // double 선언

  useEffect(() => { // count 값 변경시마다 useEffect 실행
    setDouble(count * 2); // doubl은 count의 2배
  }, [count]);

  return (
    <div>
      <p>원래 값: {count}</p>
      <p>2배 값: {double}</p>
      /* 버튼 클릭 시 setCount실행하여 count 1 증가 */
      <button onClick={() => setCount(count + 1)}>증가</button>
    </div>
  );
}

export default App;
```

```python
%%writefile /content/bootcamp/src/App.jsx

import { useState, useEffect } from 'react';

function App() {
  const [time, setTime] = useState('');
  const [remainTime, setRemainTime] = useState('');
  const [color, setColor] = useState('black');

useEffect(() => {
  const timer = setInterval(() => {
    const now = new Date();

    const target = new Date();
    target.setHours(17, 50, 0, 0);

    // 이미 지났으면 내일로
    if (target.getTime() < now.getTime()) {
      target.setDate(target.getDate() + 1);
    }

    const nowTime = now.toLocaleTimeString();
    const remainingTime = target.getTime() - now.getTime();

    // ms → 시/분/초 변환
    const hours = Math.floor(remainingTime / (1000 * 60 * 60));
    const minutes = Math.floor((remainingTime % (1000 * 60 * 60)) / (1000 * 60));
    const seconds = Math.floor((remainingTime % (1000 * 60)) / 1000);

    const displayTime = `${hours}시간 ${minutes}분 ${seconds}초`;

    // 초 기준 색상
    if (seconds >= 50) {
      setColor('red');
    } else {
      setColor('black');
    }

    setTime(nowTime);
    setRemainTime(displayTime);

  }, 1000);

  return () => clearInterval(timer);
}, []);

  return (
    <div>
      <div>한국경제신문 시간...</div>
      <h1 style={{ color }}>{time}</h1>
      <div></div>
      <div>퇴근까지</div>
      <h1 style={{ color }}>{remainTime}</h1>

    </div>
  );
}

export default App;
```

```python
%%writefile /content/bootcamp/src/App.jsx

import { useState } from 'react';

function App() {
  const [list, setList] = useState(['A', 'B', 'C']);

  function removeItem(index) {
    setList(list.filter((_, i) => i !== index));
  }

  return (
    <div>
      {list.map((item, i) => (
        <div key={i}>
          <span>{item}</span>
          <button onClick={() => removeItem(i)}>삭제</button>
        </div>
      ))}
    </div>
  );
}

export default App;
```

```python
%%writefile /content/bootcamp/src/App.jsx

import { useState } from 'react';

function App() {
  const [user, setUser] = useState({
    name: '홍길동',
    age: 20
  });

  function changeName() {
    setUser({ ...user, name: '김철수' });
  }

  return (
    <div>
      <p>{user.name}</p>
      <p>{user.age}</p>
      <button onClick={changeName}>이름 변경</button>
    </div>
  );
}

export default App;
```
