```python
%%writefile /content/node-server/server.js

const express = require('express');
const path = require('path');

const app = express();
const PORT = 3000;

// Ï†ïÏ†Å ÌååÏùº Ï†úÍ≥µ
app.use('/static', express.static(path.join(__dirname, 'static')));

// Î£®Ìä∏ Ï†ëÏÜç Ïãú React index.html Î∞òÌôò
app.get('/', (req, res) => {
  res.sendFile(path.join(__dirname, 'static', 'index.html'));
});

app.listen(PORT, () => {
  console.log(`ÏÑúÎ≤Ñ Ïã§Ìñâ: http://localhost:${PORT}`);
});
```
