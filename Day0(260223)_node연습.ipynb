```python
%%writefile server.js

// Express 모듈을 불러옵니다 (Node.js의 웹 프레임워크)
const express = require("express");

// Express 애플리케이션 인스턴스를 생성합니다
const app = express();

// 미들웨어: POST 요청의 URL-encoded 형식의 body 데이터를 파싱합니다
// extended: true 옵션은 중첩된 객체도 파싱할 수 있게 해줍니다
app.use(express.urlencoded({ extended: true }));

// GET 요청 라우트: 루트 경로("/")에 접속하면 HTML 폼을 보여줍니다
app.get("/", (req, res) => {
  // HTML 응답을 전송합니다
  // 사용자가 username을 입력할 수 있는 폼을 제공합니다
  res.send(`
    <html>
      <body>
        <h2>HTTP Request Demo</h2>
        <!-- POST 메서드로 /submit 경로에 데이터를 전송하는 폼 -->
        <form method="POST" action="/submit">
          <!-- username이라는 이름의 텍스트 입력 필드 -->
          <input type="text" name="username" placeholder="Username" />
          <!-- 폼 제출 버튼 -->
          <button type="submit">Submit</button>
        </form>
      </body>
    </html>
  `);
});

// POST 요청 라우트: 폼이 제출되면 이 핸들러가 실행됩니다
app.post("/submit", (req, res) => {
  // 서버 콘솔에 HTTP 요청 정보를 출력합니다

  console.log("===== REQUEST LINE =====");
  // HTTP 메서드(POST), 요청 URL(/submit), HTTP 버전을 출력
  console.log(req.method, req.originalUrl, req.httpVersion);

  console.log("\\n===== HEADERS =====");
  // 요청 헤더 정보를 출력 (Content-Type, User-Agent 등)
  console.log(req.headers);

  console.log("\\n===== BODY =====");
  // 폼에서 전송된 데이터(username)를 출력
  console.log(req.body);

  // 사용자에게 응답 HTML을 전송합니다
  // 입력한 username을 화면에 표시합니다
  res.send(`
    <html>
      <body>
        <h3>Form Received</h3>
        <!-- req.body.username으로 폼에서 전송된 값에 접근 -->
        <p>Username: ${req.body.username}</p>
      </body>
    </html>
  `);
});

// 서버를 3000번 포트에서 시작합니다
// 서버가 성공적으로 시작되면 콜백 함수가 실행됩니다
app.listen(3000, () => {
  console.log("Server running on port 3000");
});
```
