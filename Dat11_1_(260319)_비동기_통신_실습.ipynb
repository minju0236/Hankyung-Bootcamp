# Node.js 비동기 통신으로 MariaDB 상품 재고 현황 출력하기

```python
%%writefile server.js

const express = require("express");
const path = require("path");
const mysql = require("mysql2/promise");

const app = express();

const pool = mysql.createPool({
    host: "localhost",
    user: "testuser",
    password: "1234",
    database: "testdb",
    // 동시에 여러 사용자가 접속할 수 있기 때문에
    // DB 연결을 매번 새로 생성하지 않고 "Connection Pool"을 사용한다. Pool로 연결하지 않으면 요청하다 새로 연결해야 한다.
    // 아래 설정들은 이 풀의 동작 방식을 제어하는 옵션이다.
    waitForConnections: true,
    // 사용 가능한 DB 연결이 모두 사용 중일 때
    // 새 요청을 즉시 실패시키지 않고 "대기 큐"에서 기다리도록 설정
    // true  → 연결이 반환될 때까지 요청을 대기
    // false → 연결이 없으면 즉시 오류 발생
    connectionLimit: 10,
    // 동시에 생성할 수 있는 최대 DB 연결 수
    // 여기서는 최대 10개의 DB 연결을 유지
    // 즉, 동시에 10개의 SQL 요청을 처리 가능
    queueLimit: 0
    // 연결이 모두 사용 중일 때 대기할 수 있는 요청 수
    // 0은 "무제한 대기"를 의미
    // 예: connectionLimit이 10이고 요청이 100개면
    //     10개는 실행, 나머지 90개는 큐에서 대기
});

// 메인 페이지
app.get("/", (req, res) => {
    res.sendFile(path.join(__dirname, "templates", "index.html"));
});

// 상품 재고 API
app.get("/api/product", async (req, res) => {
    try {
        const sql = `
            SELECT id, name, category, stock, price
            FROM product
            ORDER BY id ASC
        `;

        const [rows] = await pool.query(sql);

        res.json(rows);
    } catch (error) {
        console.error("DB 조회 오류:", error);
        res.status(500).json({ message: "데이터 조회 실패" });
    }
});

app.listen(3000, () => {
    console.log("server running : http://localhost:3000");
});
```

# Node.js 비동기 통신으로 form 데이터를 MariaDB에 저장하고 상품 재고 현황 출력하기

```python
%%writefile server.js

const express = require("express");
const path = require("path");
const mysql = require("mysql2/promise");

const app = express();

app.use(express.urlencoded({ extended: true }));
app.use(express.json());

const pool = mysql.createPool({
    host: "localhost",
    user: "testuser",
    password: "1234",
    database: "testdb",
    waitForConnections: true,
    connectionLimit: 10,
    queueLimit: 0
});

// 메인 페이지
app.get("/", (req, res) => {
    res.sendFile(path.join(__dirname, "templates", "index.html"));
});

// 상품 재고 조회 API
app.get("/api/product", async (req, res) => {
    try {
        const sql = `
            SELECT id, name, category, stock, price
            FROM product
            ORDER BY id DESC
        `;

        const [rows] = await pool.query(sql);

        res.json(rows);
    } catch (error) {
        console.error("DB 조회 오류:", error);
        res.status(500).json({ message: "데이터 조회 실패" });
    }
});

// 상품 등록 API
app.post("/api/product", async (req, res) => {
    try {
        const { name, category, stock, price } = req.body;

        const sql = `
            INSERT INTO product (name, category, stock, price)
            VALUES (?, ?, ?, ?)
        `;

        await pool.query(sql, [name, category, stock, price]);

        res.json({ message: "상품 등록 완료" });
    } catch (error) {
        console.error("DB 저장 오류:", error);
        res.status(500).json({ message: "데이터 저장 실패" });
    }
});

app.listen(3000, () => {
    console.log("server running : http://localhost:3000");
});
```

# async/await 기반 MariaDB CRUD 파이프라인 구현

```python
%%writefile server.js
const express = require('express');
const mysql = require('mysql2/promise');
const path = require('path');

const app = express();
const PORT = 3000;

app.use(express.json());

const pool = mysql.createPool({
    host: 'localhost',
    user: 'testuser',
    password: '1234',
    database: 'testdb'
});

app.get('/', (req, res) => {
    res.sendFile(path.join(__dirname, 'templates', 'index.html'));
});

app.get('/employee/list', async (req, res) => {
    const [rows] = await pool.query('SELECT * FROM employee ORDER BY id DESC');
    res.json(rows);
});

app.post('/employee/add', async (req, res) => {
    const { name, dept } = req.body;

    await pool.query(
        'INSERT INTO employee (name, dept) VALUES (?, ?)',
        [name, dept]
    );

    res.json({ message: 'ok' });
});

app.put('/employee/update/:id', async (req, res) => {
    const { id } = req.params;
    const { name, dept } = req.body;

    await pool.query(
        'UPDATE employee SET name = ?, dept = ? WHERE id = ?',
        [name, dept, id]
    );

    res.json({ message: 'ok' });
});

app.delete('/employee/delete/:id', async (req, res) => {
    const { id } = req.params;

    await pool.query(
        'DELETE FROM employee WHERE id = ?',
        [id]
    );

    res.json({ message: 'ok' });
});

app.listen(PORT, () => {
    console.log(`등록 프로젝트 실행: http://localhost:${PORT}`);
});
```

```python
%%writefile templates/index.html

<!DOCTYPE html>
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <title>CRUD</title>
</head>
<body>
    <input id="name" placeholder="이름">
    <input id="dept" placeholder="부서">
    <button onclick="addEmployee()">등록</button>

    <table border="1">
        <tbody id="list"></tbody>
    </table>

    <script>
        async function loadEmployees() {
            try {
                const res = await fetch('/employee/list');
                const data = await res.json();

                const tbody = document.getElementById('list');
                tbody.innerHTML = '';

                for (const e of data) {
                    tbody.innerHTML += `
                        <tr>
                            <td>${e.id}</td>
                            <td><input id="name_${e.id}" value="${e.name}"></td>
                            <td><input id="dept_${e.id}" value="${e.dept}"></td>
                            <td><button onclick="updateEmployee(${e.id})">수정</button></td>
                            <td><button onclick="deleteEmployee(${e.id})">삭제</button></td>
                        </tr>
                    `;
                }
            } catch (err) {
                console.error(err);
                alert('조회 실패');
            }
        }

        async function addEmployee() {
            try {
                await fetch('/employee/add', {
                    method: 'POST',
                    headers: { 'Content-Type': 'application/json' },
                    body: JSON.stringify({
                        name: document.getElementById('name').value,
                        dept: document.getElementById('dept').value
                    })
                });
                loadEmployees();
            } catch (err) {
                console.error(err);
                alert('등록 실패');
            }
        }

        async function updateEmployee(id) {
            try {
                await fetch('/employee/update/' + id, {
                    method: 'PUT',
                    headers: { 'Content-Type': 'application/json' },
                    body: JSON.stringify({
                        name: document.getElementById('name_' + id).value,
                        dept: document.getElementById('dept_' + id).value
                    })
                });
                loadEmployees();
            } catch (err) {
                console.error(err);
                alert('수정 실패');
            }
        }

        async function deleteEmployee(id) {
            try {
                await fetch('/employee/delete/' + id, {
                    method: 'DELETE'
                });
                loadEmployees();
            } catch (err) {
                console.error(err);
                alert('삭제 실패');
            }
        }

        loadEmployees();
    </script>
</body>
</html>
```

# JavaScript 비동기 통신 기반 인사관리

```python
%%writefile server.js

const express = require('express');
const path = require('path');
const mysql = require('mysql2/promise');

const app = express();
const PORT = 3000;

app.use(express.json());
app.use(express.urlencoded({ extended: true }));
app.use('/static', express.static(path.join(__dirname, 'static')));

const pool = mysql.createPool({
    host: 'localhost',
    user: 'testuser',
    password: '1234',
    database: 'testdb',
    waitForConnections: true,
    connectionLimit: 10,
    queueLimit: 0,
    dateStrings: true
});

app.get('/', (req, res) => {
    res.sendFile(path.join(__dirname, 'templates', 'insert.html'));
});

app.post('/employee/create', async (req, res) => {
    try {
        const {
            emp_no,
            name,
            department,
            position,
            hire_date,
            employment_status,
            annual_salary,
            tax,
            bonus,
            address,
            phone,
            email,
            hr_note
        } = req.body;

        const sql = `
            INSERT INTO employee (
                emp_no, name, department, position, hire_date,
                employment_status, annual_salary, tax, bonus,
                address, phone, email, hr_note, created_at
            )
            VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, NOW())
        `;

        await pool.execute(sql, [
            emp_no,
            name,
            department,
            position,
            hire_date,
            employment_status,
            annual_salary,
            tax,
            bonus,
            address,
            phone,
            email,
            hr_note
        ]);

        res.json({
            success: true,
            message: '직원 정보가 정상적으로 등록되었습니다.'
        });
    } catch (error) {
        console.error('등록 실패:', error);
        res.status(500).json({
            success: false,
            message: '직원 정보 등록 중 오류가 발생했습니다.'
        });
    }
});

app.listen(PORT, () => {
    console.log(`등록 프로젝트 실행: http://localhost:${PORT}`);
});
```

```python
%%writefile templates/insert.html

<!DOCTYPE html>
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>직원 등록</title>
    <link rel="stylesheet" href="/static/style.css">
</head>
<body>
    <div class="wrap">
        <h1>직원 정보 등록</h1>

        <form id="insertForm">
            <div class="form-group">
                <label for="emp_no">사번</label>
                <input type="text" id="emp_no" required>
            </div>

            <div class="form-group">
                <label for="name">직원명</label>
                <input type="text" id="name" required>
            </div>

            <div class="form-group">
                <label for="department">부서</label>
                <input type="text" id="department" required>
            </div>

            <div class="form-group">
                <label for="position">직급</label>
                <input type="text" id="position" required>
            </div>

            <div class="form-group">
                <label for="hire_date">입사일</label>
                <input type="date" id="hire_date" required>
            </div>

            <div class="form-group">
                <label for="employment_status">재직상태</label>
                <select id="employment_status" required>
                    <option value="재직">재직</option>
                    <option value="승진">승진</option>
                    <option value="휴직">휴직</option>
                    <option value="퇴사">퇴사</option>
                </select>
            </div>

            <div class="form-group">
                <label for="annual_salary">연봉</label>
                <input type="number" id="annual_salary" required>
            </div>

            <div class="form-group">
                <label for="tax">세금</label>
                <input type="number" id="tax" required>
            </div>

            <div class="form-group">
                <label for="bonus">보너스</label>
                <input type="number" id="bonus" required>
            </div>

            <div class="form-group">
                <label for="address">주소</label>
                <input type="text" id="address" required>
            </div>

            <div class="form-group">
                <label for="phone">연락처</label>
                <input type="text" id="phone" required>
            </div>

            <div class="form-group">
                <label for="email">이메일</label>
                <input type="email" id="email" required>
            </div>

            <div class="form-group">
                <label for="hr_note">인사 메모</label>
                <textarea id="hr_note" required></textarea>
            </div>

            <button type="submit">저장하기</button>
        </form>

        <div id="resultBox" class="result-box"></div>
    </div>

    <script>
        const insertForm = document.getElementById('insertForm');
        const resultBox = document.getElementById('resultBox');

        insertForm.addEventListener('submit', async function (event) {
            event.preventDefault();

            const employeeData = {
                emp_no: document.getElementById('emp_no').value,
                name: document.getElementById('name').value,
                department: document.getElementById('department').value,
                position: document.getElementById('position').value,
                hire_date: document.getElementById('hire_date').value,
                employment_status: document.getElementById('employment_status').value,
                annual_salary: document.getElementById('annual_salary').value,
                tax: document.getElementById('tax').value,
                bonus: document.getElementById('bonus').value,
                address: document.getElementById('address').value,
                phone: document.getElementById('phone').value,
                email: document.getElementById('email').value,
                hr_note: document.getElementById('hr_note').value
            };

            try {
                const response = await fetch('/employee/create', {
                    method: 'POST',
                    headers: { 'Content-Type': 'application/json' },
                    body: JSON.stringify(employeeData)
                });

                const result = await response.json();

                resultBox.style.display = 'block';

                if (result.success) {
                    resultBox.className = 'result-box success';
                    resultBox.textContent = result.message;
                    insertForm.reset();
                } else {
                    resultBox.className = 'result-box error';
                    resultBox.textContent = result.message;
                }
            } catch (error) {
                console.error(error);
                resultBox.style.display = 'block';
                resultBox.className = 'result-box error';
                resultBox.textContent = '서버 통신 중 오류가 발생했습니다.';
            }
        });
    </script>
</body>
</html>
```
