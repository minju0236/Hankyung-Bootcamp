```python
%%writefile server.js

const express = require("express");
const path = require("path");
const app = express();

/* static íŹ´ëŤ”ëĄĽ ę·¸ëŚ€ëˇś ëŁ¨íŠ¸ëˇś ě‚¬ěš© */
app.use(express.static(path.join(__dirname, "static")));
// app.use("/static", express.static(path.join(__dirname, "static")));

const TEMPLATE_PATH = path.join(__dirname, "templates");

// Home
app.get("/", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "index.html"));
});

// About
app.get("/about", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "about.html"));
});

// Products
app.get("/products", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "products.html"));
});

// Loan
app.get("/loan", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "loan.html"));
});

// Support
app.get("/support", (req, res) => {
  res.sendFile(path.join(TEMPLATE_PATH, "support.html"));
});

app.listen(3000, () => {
  console.log("Server running");
});
```
