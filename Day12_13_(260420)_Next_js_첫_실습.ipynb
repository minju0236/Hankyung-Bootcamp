```python
%%writefile first-next-routing/components/Header.jsx

'use client';

import Link from 'next/link';
import { usePathname } from 'next/navigation';

const menuItems = [
  { label: '은행소개', href: '/about' },
  { label: '통장', href: '/account' },
  { label: '예금 · 적금', href: '/savings' },
  { label: '대출', href: '/loan' },
  { label: '외환', href: '/fx' },
  { label: '카드', href: '/card' },
  { label: '고객센터', href: '/support' },
  { label: '채용', href: '/careers' },
  { label: '기업뱅킹', href: '/business' }
];

export default function Header() {
  const pathname = usePathname();

  return (
    <header className="site-header-wrapper">
      <div className="browser-bar">
        <div className="browser-left">
          <span className="traffic light-red"></span>
          <span className="traffic light-yellow"></span>
          <span className="traffic light-green"></span>
        </div>

        <div className="browser-tab-area">
          <div className="browser-tab">토스뱅크</div>
          <div className="browser-plus">+</div>
        </div>

        <div className="browser-right"></div>
      </div>

      <div className="address-bar">
        <div className="address-left">
          <span>←</span>
          <span>↻</span>
        </div>

        <div className="address-input">tossbank.com{pathname}</div>

        <div className="address-right">
          <span>☆</span>
          <span>⬇</span>
          <span>⋮</span>
        </div>
      </div>

      <nav className="main-nav">
        <Link href="/" className="logo">
          toss bank
        </Link>

        <ul className="menu-list">
          {menuItems.map(function (item) {
            const isActive = pathname === item.href;

            return (
              <li key={item.href}>
                <Link
                  href={item.href}
                  className={isActive ? 'menu-item active-menu' : 'menu-item'}
                >
                  {item.label}
                </Link>
              </li>
            );
          })}
        </ul>

        <button type="button" className="start-button">
          시작하기
        </button>
      </nav>
    </header>
  );
}
```
