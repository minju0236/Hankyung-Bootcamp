```python
%%writefile /content/index.html

<!DOCTYPE html>
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <title>CSS 선택자 우선순위 실습</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>

    <div id="main-container" class="wrapper">
        <h1 id="title" class="heading" style="color: blue;">
            CSS 우선순위 테스트
        </h1>

        <p class="description">
            우선순위 점수에 따라 글자 색상이 결정됩니다.
        </p>

        <button class="btn active">확인 버튼</button>
    </div>

</body>
</html>
```

```python
%%writefile /content/style.css

/* 5순위: 요소 선택자 (1점) */
h1 {
    color: gray;
}

/* 4순위: 클래스 선택자 (10점) */
.heading {
    color: green;
}

/* 3순위: ID 선택자 (100점) */
#title {
    color: orange;
}

/* 2순위: 인라인 스타일 (1000점)
   -> HTML 태그 안에 작성된 style="color: blue;"가 적용됨 */

/* 1순위: !important (무적)
   -> 아래 주석을 해제하면 모든 것을 무시하고 빨간색이 됩니다. */
/*
h1 {
    color: red !important;
}
*/

/* ------------------------------------------ */

/* 코드 순서(Source Order) 테스트 */
/* 두 선택자의 점수가 10점으로 같을 때 */

.description {
    color: black;
}

.description {
    color: purple; /* 더 아래에 있으므로 보라색이 최종 적용됨 */
}

/* ------------------------------------------ */

/* 복합 선택자 점수 계산 실습 */

/* 요소(1) + 요소(1) = 2점 */
div button {
    background-color: lightgray;
}

/* 클래스(10) + 요소(1) = 11점 (우승) */
.wrapper button {
    background-color: yellow;
}

/* 클래스(10) + 가상클래스(10) = 20점 (최종 우승) */
.btn.active {
    background-color: lightgreen;
    border: 2px solid green;
}
```

```python
from IPython.display import HTML

# Read the HTML file content
with open('/content/index.html', 'r') as f:
    html_content = f.read()

# Display the HTML content
HTML(html_content)
```

```python
%%writefile /content/index.html

<!DOCTYPE html>
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <title>마진과 패딩 상세 실습</title>
    <link rel="stylesheet" href="style.css">
</head>
<body>

    <div class="container">
        <a href="#" class="button">클릭 범위 확인</a>

        <div class="box box-1">상하좌우 동일 (1개 값)</div>
        <div class="box box-2">상하/좌우 구분 (2개 값)</div>
        <div class="box box-4">시계 방향 적용 (4개 값)</div>

        <div class="box negative-margin">음수 마진으로 겹치기</div>
    </div>

</body>
</html>
```

```python
%%writefile /content/style.css

/* 박스 모델 초기화 */
* {
    margin: 0;
    padding: 0;
    box-sizing: border-box;
}

body {
    padding: 50px;
    background-color: #f0f0f0;
}

/* 1. 패딩과 클릭 영역 */
.button {
    display: inline-block;
    background-color: #3498db;
    color: white;
    text-decoration: none;
    /* 패딩을 통해 글자 주변의 클릭 가능 영역을 확장 */
    padding: 15px 30px;
    border-radius: 5px;
    margin-bottom: 30px;
}

/* 2. 단축 속성(Shorthand) 실습 */
.box {
    background-color: white;
    border: 2px solid #333;
    width: 300px;
    font-size: 14px;
}

/* 값 1개: 상하좌우 30px 동일 */
.box-1 {
    padding: 30px;
    margin: 20px;
}

/* 값 2개: 상하 10px, 좌우 50px */
.box-2 {
    padding: 10px 50px;
    margin: 20px 0;
}

/* 값 4개: 상(0), 우(50px), 하(20px), 좌(10px) - 시계 방향 */
.box-4 {
    padding: 0 50px 20px 10px;
    margin: 20px 0;
}

/* 3. 음수 마진과 중앙 정렬 */
.negative-margin {
    background-color: #e74c3c;
    color: white;
    /* 마진 0 auto는 블록 요소를 중앙으로 보냄 */
    margin: 0 auto;
    /* 음수 마진을 사용하여 위쪽 요소와 겹치게 설정 */
    margin-top: -30px;
    padding: 20px;
}
```
