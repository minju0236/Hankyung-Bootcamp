```python
%%writefile server.js

const express = require("express");
const path = require("path");

const app = express();
const PORT = 3000;

const products = [
    { id: 1, name: "MacBook Pro", price: 2890000, category: "Laptop", stock: 12 },
    { id: 2, name: "Mechanical Keyboard", price: 189000, category: "Accessory", stock: 30 },
    { id: 3, name: "4K Monitor", price: 620000, category: "Display", stock: 8 },
    { id: 4, name: "Wireless Mouse", price: 99000, category: "Accessory", stock: 24 }
];

app.use("/static", express.static(path.join(__dirname, "static")));

app.get("/", (req, res) => {
    res.sendFile(path.join(__dirname, "templates", "index.html"));
});

app.get("/api/products", (req, res) => {
    setTimeout(() => {
        res.json(products);
    }, 700);
});

app.listen(PORT, () => {
    console.log(`server running on http://localhost:${PORT}`);
});
```

```python
%%writefile templates/index.html

<!DOCTYPE html>
<html lang="ko">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>상품 대시보드</title>
    <link rel="stylesheet" href="/static/style.css">
</head>
<body>
    <div class="page">
        <header class="hero">
            <div>
                <p class="eyebrow">Inventory Dashboard</p>
                <h1>상품 목록 조회</h1>
                <p class="hero-text">비동기 GET 요청으로 상품 데이터를 불러와 카드형 UI로 출력합니다.</p>
            </div>
            <button id="loadBtn" class="primary-btn">상품 불러오기</button>
        </header>

        <section id="statusBox" class="status-box">대기 중</section>

        <section id="productGrid" class="grid"></section>
    </div>

    <script>
        const loadBtn = document.getElementById("loadBtn");
        const statusBox = document.getElementById("statusBox");
        const productGrid = document.getElementById("productGrid");

        loadBtn.addEventListener("click", async function () {
            statusBox.textContent = "서버에 상품 목록 요청 중...";
            productGrid.innerHTML = "";

            try {
                const response = await fetch("/api/products");
                const data = await response.json();

                statusBox.textContent = "상품 목록 조회 완료";

                data.forEach(function (item) {
                    const card = document.createElement("article");
                    card.className = "card";

                    card.innerHTML = `
                        <div class="card-top">
                            <span class="badge">${item.category}</span>
                            <span class="stock">재고 ${item.stock}</span>
                        </div>
                        <h3>${item.name}</h3>
                        <p class="price">${item.price.toLocaleString("ko-KR")}원</p>
                        <p class="meta">상품번호: ${item.id}</p>
                    `;

                    productGrid.appendChild(card);
                });
            } catch (error) {
                statusBox.textContent = "조회 실패";
                console.error(error);
            }
        });
    </script>
</body>
</html>
```
